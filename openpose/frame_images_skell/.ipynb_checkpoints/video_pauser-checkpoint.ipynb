```python
import cv2
```

```python
vidcap = cv2.VideoCapture('../output.avi')
success,image = vidcap.read()
```

```python
count = 0
while success:
  cv2.imwrite("../frame_images/"+"frame%d.jpg" % count, image)     # save frame as JPEG file      
  success,image = vidcap.read()
  print('Read a new frame: ', success)
  count += 1
```
